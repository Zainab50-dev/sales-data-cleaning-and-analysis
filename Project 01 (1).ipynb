```python
# Sales Data Cleaning and Analysis
```

```python
import pandas as pd
```

```python
# Loading Dataset
```

```python
df = pd.read_csv('sales_data_sample.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
Column_names = df.columns
Column_names
```

```python
# Handling Missing Values
```

```python
Missing_values = df.isnull().sum()
Missing_values
```

```python
df['TERRITORY']= df['TERRITORY'].fillna('Unknown Region')
```

```python
df['STATE']= df['STATE'].fillna('Unknown')
```

```python
# Dropping Irrelevant Columns
```

```python
Dropping_columns = df.drop(columns = ['QTR_ID','MONTH_ID','YEAR_ID','MSRP','CUSTOMERNAME','PHONE','ADDRESSLINE1', 'ADDRESSLINE2','POSTALCODE','CONTACTLASTNAME', 'CONTACTFIRSTNAME'],inplace = True)
```

```python
# Converting ORDERDATE to Datetime and creating Month/Year columns
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
df['Month'] = df['ORDERDATE'].dt.month
df['Year'] = df['ORDERDATE'].dt.year
```

```python
## Insights
```

```python
# Monthly Sales
```

```python
monthly_sales = df.groupby(['Year','Month'])['SALES'].sum().reset_index()
monthly_sales
```

```python
# Yearly Sales
```

```python
yearly_sales = df.groupby('Year')['SALES'].sum().reset_index().sort_values(by = 'SALES',ascending = False)
yearly_sales
```

```python
# City-wise Sales
```

```python
city_sales = df.groupby('CITY')['SALES'].sum().reset_index().sort_values(by = 'SALES',ascending = False)
city_sales
```

```python
# Top Product Lines
```

```python
top_product_lines = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending = False).head(5)
top_product_lines
```

```python
df.to_csv('cleaned_sales_data.csv')
```
